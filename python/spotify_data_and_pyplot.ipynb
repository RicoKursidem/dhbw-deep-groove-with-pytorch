```python
import matplotlib.pyplot as plt
import numpy as np

ms = df['duration_ms'].mean()

print("The average duration of a Song is {:.0f} ms. Or {:.3f} minutes.".format(ms, ms / 60000))
```

```python
df.isnull().sum()
```

```python
df['explicit'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%')
```

only a smal portion of the training set is explicit. That means that a great portion of the Tracks in the training set are availeble outside of Spotify.

```python
plt.hist(df['popularity'],bins=100)
plt.xlabel("popularity")
plt.ylabel("Number of Songs")
plt.show()
```

There are a lot of songs with no popularity and very vew with nearly 100%.

```python
plt.hist(df['danceability'],bins=100)
plt.xlabel("dancebility")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.scatter(df['danceability'], df['popularity'], s=0.1)
plt.xlabel("danceability")
plt.ylabel("popularity")
plt.show()
```

```python
data = df.loc[df['track_genre'] == 'rock']

plt.scatter(data['danceability'], data['popularity'], s=1)
plt.xlabel("danceability")
plt.ylabel("popularity")
plt.show()
```

dancebility and popularity seem not to coralate that much in rock.
